```python
df["OverTime"].replace({"Y":1,"N":0}, inplace = True)
df["OverTime"].replace({"yes":1,"no":0}, inplace = True)
df["OverTime"].replace({"Yes":1,"No":0}, inplace = True)
```

```python
df["Attrition"].replace({"Y":1,"N":0}, inplace = True)
df["Attrition"].replace({"yes":1,"no":0}, inplace = True)
df["Attrition"].replace({"Yes":1,"No":0}, inplace = True)
```

```python
df['JobSatisfaction'] = df['JobSatisfaction'].str.replace(",","").astype(int)
```

```python
df['Age'] = df['Age'].str.replace(",","").astype(int)
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

ohe = OneHotEncoder()
ohe.fit(df[["Department"]])

ct = make_column_transformer((OneHotEncoder(categories=ohe.categories_),["Department"]), remainder = "passthrough")
x = df.drop("Attrition", axis = 1)
y = df["Attrition"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.model_selection import train_test_split 
```

```python
lr = LogisticRegression()
pipe = make_pipeline(ct,lr)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
pipe.fit(x_train, y_train) 
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
